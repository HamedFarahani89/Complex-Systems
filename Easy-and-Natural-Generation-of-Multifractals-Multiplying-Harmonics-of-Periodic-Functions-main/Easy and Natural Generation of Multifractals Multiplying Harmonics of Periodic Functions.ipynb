```python
import numpy as np
from random import*
import matplotlib.pyplot as plot

time= np.arange(0, 1, 0.01);

amplitude   = []
for i in time:
    amplitude.append(1-(0.25*(1-np.cos((2*np.pi*i)))**2))
plot.figure(figsize=(10,10))
plot.ylabel('w(x)')
plot.xlabel('x')
plot.xlim(0,1)
plot.ylim(0,1)    
plot.plot(time, amplitude)
```

```python
import numpy as np
from random import*
import matplotlib.pyplot as plot

time= np.arange(0, 1, 0.001);

A=[]
b=2;n=5
for i in time:
    amplitude   = []
    for j in range(1,n):
        amplitude.append(1-(0.25*(1-np.cos(((b**(j-1))*2*np.pi*i)))**2))
    A.append(prod(amplitude))
plot.figure(figsize=(10,10)) 
plot.ylabel('$m^2_5(x)$')
plot.xlabel('x')
plot.xlim(0,1)
plot.ylim(0,1)
plot.plot(time, A)
```
