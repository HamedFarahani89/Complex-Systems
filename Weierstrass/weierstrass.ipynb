```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def weierstrass(t,D):
    y=[]
    tt=[]
    for i in np.arange(0,t,0.01):
        x = []
        tt.append(i)
        for n in range(0,400):
            x.append(((np.cos((2**n)*i))/(2**((n)*(2-D)))))
        y.append(sum(x))
    plt.plot(tt,y)
```

```python
plt.subplot(221)
weierstrass(2*np.pi,1.2)
plt.subplot(222)
weierstrass(2*np.pi,1.4)
plt.subplot(223)
weierstrass(2*np.pi,1.5)
plt.subplot(224)
weierstrass(2*np.pi,1.8)
plt.figure(figsize=(5,5))
plt.suptitle('Categorical Plotting of various D')
plt.show()
```
